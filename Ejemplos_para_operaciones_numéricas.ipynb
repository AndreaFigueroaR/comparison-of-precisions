```julia
# Distintas formas de mostrar mensajes y valores de variables

println("El tipo de dato de pi es $(typeof(pi))")
show(pi)
display(pi)

display(convert(Float16, pi))
# f0 al final indica el tratamiento como Float32

display(convert(Float32, pi))
display(convert(Float64, pi))
```

```julia
#Formatos de salida

x = 9.2654
y = 7.1625
suma = x + y
display(suma)
display(round(suma))
display(round(suma, digits = 3))
display(round(suma, RoundNearest))
display(round(suma, RoundUp))
display(round(suma, RoundDown))
display(round(suma, RoundToZero))
display(round(suma, RoundFromZero))
```

```julia
# Suma de números distintos en magnitud

a = 0.002
b = 600

suma_Float16 = Float16(a) + Float16(b)
suma_Float32 = Float32(a) + Float32(b)
suma_Float64 = Float64(a) + Float64(b)

print("La suma en Float16 es: ", suma_Float16)
print("\nLa suma en Float32 es: ", suma_Float32)
print("\nLa suma en Float64 es: ", suma_Float64)
```

```julia
# Suma con propagación de error

suma = 0
for i in 1:10000
    suma = Float32(Float32(suma) + Float32(0.0001))
end
print("La suma es: ", suma)
```

```julia
# Resta de números "casi" iguales

a = 0.2145
b = 0.2144
resta = a - b
print("La resta es: ", resta)
```

```julia
# Overflow para el producto

a = 0.5e10+8
b = 0.2e10+9
producto = Float16(Float16(a) * Float16(b))
print("El producto es: ", producto)
```

```julia
# Overflow para el cociente

a = 2000000
b = 0.000005
cociente = Float16(Float16(a) / Float16(b))
print("El producto es: ", cociente)
```

```julia
#Cálculo del epsilon de máquina

eps_maquina = 1.0
k = 1
continuar = true

while continuar == true

  println("Iteración: $k eps: $eps_maquina")
  eps_maquina = 0.5 * eps_maquina
  suma = 1.0 + eps_maquina

  if suma == 1.0
     continuar = false
  else
    k = k + 1
  end
end
println("eps por Julia: $(eps())")
```
